```python
def load_dataset(csv_file_path="annual-enterprise-survey-2023-financial-year-provisional.csv"):
    # Import the data
    df = pd.read_csv("annual-enterprise-survey-2023-financial-year-provisional.csv")
    print(df.head(5))

    # Display basic dataset information

    print("n1/. Number of Rows and Columns:")
    print(df.shape)

    print("\n2. First Five Rows:")
    print(df.head())

    print("\n3. Number of Missing Values per Column:")
    print(df.isnull().sum())

    print("\n4. Summary of Numerical Columns:")
    print("Sum of values:")
    print(df.select_dtypes(include='number').sum())

    print("\nAverage of values:")
    print(df.select_dtypes(include='number').mean())

    print("\nMinimum values:")
    print(df.select_dtypes(include='number').min())

    print("\nMaximum values:")
    print(df.select_dtypes(include='number').max())

    return df

 # Function to export the dataset
def export_dataset(df, file_path):
        df.to_csv(file_path, index=False)
        print(f"\nDataset exported successfully to {file_path}")

# Example usage
if __name__ == "__main__":
    # Replace 'your_file.csv' with the uploaded file name after uploading
    input_file = 'annual-enterprise-survey-2023-financial-year-provisional.csv'

    # Load and process dataset
    df = load_dataset(input_file)

    # Export the dataset (optional)
    output_file = 'output_data.csv'
    export_dataset(df, output_file)
```
